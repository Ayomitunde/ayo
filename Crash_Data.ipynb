```python
crash_dummy['Hour'] = crash_dummy['Crash Time'].str[:2]
# function for time of day
def time_of_day(hour):
    if hour >= str('5') and hour < str('12'):
        return "Morning"
    elif hour >= str('12') and hour < str('16'):
        return "Afternoon"
    elif hour >= str('12') and hour < str('19'):
        return "Evening"
    elif hour >= str('19') and hour < str('23'):
        return "Night"
    else:
        return "Midnight"

crash_dummy['Time'] = crash_dummy['Hour'].apply(time_of_day)
crash_dummy[['Crash Time','Hour', 'Time']]
```

```python
# What time category do crashes happen the most
day_groups = ['Morning', 'Afternoon', 'Evening', 'Afternoon', 'Midnight']
time_accidents = crash_dummy.groupby('Time').size().reindex(day_groups)
time_accidents
```

```python
# Average accidents per time group
crash_hourly = crash_dummy.set_index('Crash Date').resample('H')['Report Number'].size().reset_index()
crash_hourly.columns = ['Date', 'Count']
crash_hourly['Day_Name'] = crash_hourly['Date'].dt.day_name()
crash_hourtime = crash_hourly.groupby('Day_Name')['Count'].mean().reset_index()
crash_hourtime.columns = ['Day', 'Average Accidnets']
crash_hourtime.set_index('Day', inplace = True)
crash_hourtime
```

```python
days = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']
fig,ax = plt.subplots(figsize = (10, 7))
crash_hourtime.reindex(days).plot(kind = 'barh', ax = ax)
ax.set_title('\nThe average accidents that happen per hour for each day of week')
ax.set(ylabel = '\nDays', xlabel = '\nCount of accidents')
sns.despine(left=True, right=True, bottom=True, top=True)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
crash_weekly = crash_dummy.set_index('Crash Date').resample('1d')['Report Number'].size().reset_index()
crash_weekly.columns = ['Date', 'Count']
crash_weekly['Day'] = crash_weekly['Date'].dt.day_name()
weektime = crash_weekly.groupby('Day')['Count'].mean().reset_index()
weektime.columns = ['Day', 'Average Accidents']
weektime.set_index('Day', inplace = True)
weektime
```

```python
days = ['Sunday', 'Saturday', 'Friday', 'Thursday', 'Wednesday', 'Tuesday', 'Monday']
fig,ax = plt.subplots(figsize = (10, 7))
weektime.reindex(days).plot(kind = 'barh', ax = ax)
ax.set_title('\nThe average accidents per day')
ax.set(ylabel = '\nDays', xlabel = '\nCount of accidents')
sns.despine(left=True, right=True, bottom=True, top=True)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
fig,ax = plt.subplots(figsize = (10,5))
crash_dummy.set_index('Crash Date').resample('M').size().plot(label = 'Total Crashes per Month', ax = ax)
crash_dummy.set_index('Crash Date').resample('M').size().rolling(window = 12).mean().plot(color = 'darkred', linewidth = 5, label = '12 month moving average', ax = ax)
ax.set_title("Crashes per month", fontsize = 14, fontweight = 'bold')
ax.set(ylabel='Total Count\n')
ax.legend(bbox_to_anchor =(1.1,1.1), frameon = False)
sns.despine(ax = ax, top = True, right = True, left = True, bottom = False)
```

```python
# Collision Type of the crashes
collision_type = crash_dummy['Collision Type'].value_counts().reset_index()
collision_type.rename(columns={'index':'Collision Type', 'Collision Type': 'Count'}, inplace=True)

fig, ax = plt.subplots(figsize = (10, 12))
sns.barplot(y = 'Collision Type', x = 'Count', data = collision_type, ax = ax, palette='mako_r')
ax.set_title('\nCollision Type of the crashes')
ax.set(xlabel='\nCollision Type', ylabel='\nTotal Count')
sns.despine(left=True, right=True, top=True, bottom=True)
```

```python
surface_condition = crash_dummy.groupby(['Collision Type','Surface Condition']).size().reset_index(name = 'Count')
surface_condition = surface_condition.sort_values(by='Count', ascending=False)
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Collision Type', y = 'Count', hue='Surface Condition', data=surface_condition.head(20), palette='gist_heat')
plt.title("Collison vs Surface Condition")
ax.legend(bbox_to_anchor=(1, .5), loc='center left')
sns.despine()
plt.tight_layout()
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```
